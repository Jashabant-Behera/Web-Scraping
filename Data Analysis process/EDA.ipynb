## Exploratory Data Analysis

### Why do EDA?

* `Model Building` : Creating predection model
* `Analysis and Reporting` : To extract KPI or Knowledgable information
* `Validating Assumptions` : To test the assumptions
* `Handeling Missing Values` 
* `Feature Engineering` : Creating new columns that help build model or to make changes to the existing columns
* `Detecting Outliers`

### Column Types
* Numerical
* Categorical 
* Mixed

### Steps in EDA
1. `Label the columns` : Numerical, Categorical, Mixed
2. `Univariate Analysis` : Analysis of single column
3. `Bivariate Analysis` : Analysis of two columns
4. `Multivariate Analysis` : Analysis of more than two columns
5. `Feature Engineering` : Creating new columns that help build model or to make changes to the existing columns
 

### 2. Univariate Analysis:
Univariate analysis focuses on analyzing each feature in the dataset independently.

* `Distribution Analysis`: The distribution of each feature is examined to identify its shape, central tendency, and dispersion

* `Identifying Potential Issues`: Univariate analysis helps identify potential issues such as missing values, outliers, or skewed distributions that may require further investigation or preprocessing.

The shape of data distribution refers to ots overall pattern or form as it is represnted on a graph. Some common shapes of data distribution include:
- `Normal Distribution`: Symmetrical bell-shaped curve where most data points cluster around the mean.
- `Skewed Distribution`: Asymmetrical distribution where data points are concentrated on one side, resulting in a longer tail on the other side. It can be positively skewed (right-skewed) or negatively skewed (left-skewed).
- `Uniform Distribution`: All values within a certain range have an equal probability of occurring, resulting in a flat or rectangular shape.
- `Bimodal Distribution`: Distribution with two distinct peaks or modes, indicating the presence of two different groups or populations within the data.

### Steps of doing Univeriate Analysis:
1. `Descriptive Statistics` : Mean, Median, Mode, Standard Deviation, Variance, Percentiles
2. `Data Visualization` : Histograms, Box Plots, Density Plots, Bar Charts
3. `Identifying Outliers` : Using statistical methods (e.g., Z-score, IQR) or visualizations (e.g., box plots) to identify potential outliers in the data. It is imprtant to determine whether to keep, transform, or remove outliers based on their impact on the analysis.
4. `Skewness and Kurtosis` : Calculating skewness and kurtosis to understand the asymmetry and peakedness of the data distribution using statistical measures or methods.
5. `Conclusion` : Summarizing the findings from the univariate analysis and identifying any potential issues or areas for further investigation.

### Steps of doing Bivarient Analysis:
* Select 2 Columns for analysis
* Understand type of relationship:
    1. Numerical - Numerical
        * You can plot graphs like Scatterplot(regression plots), 2D histplot, 2D KDEplots
        * Check correlation coefficient to check linear relationship
    2. Numerical - Categorical: Create visualizations that compare the distribution of the numerical data across different categories of the categorical data
        * Use boxplots, violin plots, or strip plots
    3. Categorical - Categorical: 
        * You can create cross-tabulations or contingency tables that show the distribution of values in one categorical column, grouped by the values in the other categorical column.
        * You can plot like heatmaps, stacked barplots, treemaps
* Write your conclusions
